## 一、数据的获取
    可以使用sklearn内部数据集、计算机生成的数据集、网络上的数据集、文件中的数据集.在这个一般流程中我们使用sklearn内部自带数据集作为输入。


```python
#引入三种python数据分析常用工具包
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets   #可以理解为从包中调用具体对象datasets

iris=datasets.load_iris()   #加载鸢尾花数据集
```

```python
iris    #显示鸢尾花数据集的全部信息
```

```python
iris.data    #显示鸢尾花数据集的输入数据
```

```python
iris.target#显示鸢尾花数据集的标签数据
```

```python
iris.feature_names   #显示鸢尾花数据的输入特征名称
```

```python
iris.target_names
```

```python
#利用dataframe进行简单的可视化分析
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df.plot(figsize=(15,15))
```

## 二、数据预处理
 预处理数据包括：数据的标准化、数据的归一化、数据的二值化、非线性变换、数据特征编码、处理缺失值等。

```python
from sklearn import preprocessing  #从sklearn这个包中导入preprossing（可以理解为c++里面的一个类）
```

```python
scaler=preprocessing.MinMaxScaler()#将样本特征值线性缩放在0,1之间,最大的记为1，最小的记为0
```

```python
data=scaler.fit_transform(iris.data)
```

```python
target=iris.target
```

```python
data
```

## 三、模型的训练
  根据问题的特点选择合适的估计器estimater模型

```python
from sklearn.model_selection import train_test_split   #导入切分训练集和测试集的函数
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.3)#切分训练集和测试集
len(X_train),len(y_test)
```

```python
from sklearn import svm
```

```python
mod=svm.SVC(kernel='linear',C=1,probability=True)
mod.fit(X_train,y_train)
mod.predict(X_test)
```

```python
mod.get_params()   #显示模型的参数设置
```

```python
mod.predict_proba(X_test)#得到模型的预测精度
```

```python
mod.score(X_test,y_test)
```

## 四、模型的评估
模型的评估和模型的优化功能是在sklearn.model_selection中，除了使用mod.score（x_test,y_test）简单粗暴地评估模型的质量之外，可以使用sklearn.metrics模块针对不同的问题类型提供了各种评估指标并且可以创建用户自定义的评估指标。可以使用交叉验证评估模型的泛化能力，能够有效避免过拟合。

```python
from sklearn.metrics import classification_report
#使用sklearn.metrics模块内的评价指标
print(classification_report(target,mod.predict(data),target_names=iris.target_names))
```

```python
from sklearn.model_selection import cross_val_score
#使用sklearn.model_selection里面的cross_val_score()得到每次验证的平均得分
scores=cross_val_score(mod,data,target,cv=5) 
```

```python
print(scores)
```

```python
print("Accuracy: %0.2f(+/-%0.2f)"%(scores.mean(),scores.std()*2)) #输出平均得分和95%置信区间
```

## 五、模型的优化
优化模型的方法主要有：网格搜索算法、随机搜索算法、模型特定交叉验证、信息准则优化。
网格搜索法指的是对于超参数空间的每一种可能情况进行交叉验证评分评选出最好的超参数组合。

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
```

```python
mod=svm.SVC()
```

```python
#定义超参数空间
param_grid=[{'C':[0.1,1,10,100,1000],'kernel':['linear']},
            {'C':[0.1,1,10,100,1000],'gamma':[0.001,0.1,1],'kernel':['rbf']}]
```

```python
clf=GridSearchCV(mod,param_grid,scoring='accuracy',cv=5)  #调用网格搜索并进行相关参数设置
clf.fit(data,target)
clf.predict(data)   #优化了的分类器进行分类
```

```python
clf.get_params()#查看全部参数
```

```python
clf.best_params_#查看最优参数
```

```python
clf.best_score_
```

## 六、模型的持久化
1.可以使用python的内置pickle模块将训练好的模型保存到磁盘或者字符串；<br>
2.使用sklearn里面的joblib会更有效，但只能保存到磁盘；

```python
import pickle
s=pickle.dumps(clf)  #保存模型成字符串
clf1=pickle.loads(s)  #从字符串加载模型
clf1.predict(data)
```

```python
from sklearn.externals import joblib
```

```python
joblib.dump(clf,'clf')
```

```python
clf2=joblib.load('clf')
clf2.predict(data)
```
