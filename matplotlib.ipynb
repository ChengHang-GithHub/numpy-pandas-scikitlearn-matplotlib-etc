```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
data=np.arange(10)
plt.plot(data)
```

### 9.1.1 图片与子图
matplotlib所绘制的图位于图片（figure）对象中。可以使用plt.figure生成一个新的图片。

```python
fig=plt.figure()
#建立三个子图对象实现面向对象的的编程
ax1=fig.add_subplot(2,2,1)#前两个参数代表图片内子图的规格（2*2），最后一个参数表示子图在图片内的序号
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax1.plot(np.arange(50),np.arange(50).cumsum(),'k--')
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
ax3.hist(np.random.randn(100),bins=20,color='r',alpha=0.8)
```

```python
#此方法与上面的方法等效
fig,axes=plt.subplots(2,2)
axes[0,0].plot(np.arange(50),np.arange(50).cumsum(),'k--')
axes[0,1].scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
axes[1,0].hist(np.random.randn(100),bins=20,color='r',alpha=0.8)
```

#### 9.1.1.1 调整子图周围的间距

```python
fig,axes=plt.subplots(2,2)
axes[0,0].plot(np.arange(50).cumsum(),'k--')
axes[0,1].scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
axes[1,0].hist(np.random.randn(100),bins=20,color='r',alpha=0.8)
plt.subplots_adjust(wspace=0.2,hspace=0.2)#设置子图之间的间距
```

## 9.1.2颜色、标记和线类型
matplotlib的主函数plot接收带有x和y轴的数组以及一些可选的字符串缩写参数来指明颜色和线类型。

```python
fig,axes=plt.subplots(2,2)
axes[0,0].plot(np.arange(5).cumsum(),linestyle=':',color='b',marker='o')#设置子图颜色，标记和线类型
axes[0,1].scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
axes[1,0].hist(np.random.randn(100),bins=20,color='r',alpha=0.8)
plt.subplots_adjust(wspace=0.2,hspace=0.2)
```

### 9.1.3 刻度、标签和图例
#### 9.1.3.1 设置标题、轴标签、刻度和刻度标签
ax.set_xticks()<br>
ax.set_xticklabels()<br>
ax.set_title()<br>
ax.set_xlabel()<br>
注：ax为所建子图对象！

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0,250,500,750,1000]) #设置x轴刻度
ax.set_xticklabels(['one','two','three','four','five'],rotation=45,fontsize='small')#设置x轴刻度标签
ax.set_title('my first matplotlib plot')#设置标题
ax.set_xlabel('stage')  #设置x轴标签
```

#### 9.1.3.2 添加图例
ax.legend('best')实现图例添加！

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1) 
ax.plot(np.random.randn(1000).cumsum(),linestyle='-.',color='k',label= 'one')
ax.plot(np.random.randn(1000).cumsum(),linestyle='--',color='r',label= 'two')
ax.plot(np.random.randn(1000).cumsum(),linestyle=':',color='b',label= 'three')
ax.legend('best')  #添加图例
```

### 9.1.4注释与子图加工
除了标准的绘图类型，你可能还会想在图表上绘制自己的注释，而且注释可能会包含文本、箭头以及其他图形。你可以使用text、arrow、和annote方法来添加注释和文本。

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),linestyle='-.',color='k',label= 'one')
ax.plot(np.random.randn(1000).cumsum(),linestyle='--',color='r',label= 'two')
ax.plot(np.random.randn(1000).cumsum(),linestyle=':',color='b',label= 'three')
ax.set_xticks([0,250,500,750,1000]) #设置x轴刻度
ax.set_xticklabels(['one','two','three','four','five'],rotation=45,fontsize='small')#设置x轴刻度标签
ax.set_title('my first matplotlib plot')#设置标题
ax.set_xlabel('stage')  #设置x轴标签
ax.legend('best')  #添加图例
ax.text(100,20,'hello,world!',fontsize=10)#使用text（）在指定位置绘制文本

'''
    xy=(横坐标，纵坐标)  箭头尖端
    xytext=(横坐标，纵坐标) 文字的坐标，指的是最左边的坐标
    arrowprops= {
        facecolor= '颜色',
        shrink = '数字' <1  收缩箭头
    }
'''

data=[
    (100,'aaaaaa'),
    (200,'bbbbbb'),
    (300,'cccccc')
]
for data,label in data:
    ax.annotate(label,xy=(data,10),xytext=(data,15),   #使用annotate（）方法
               arrowprops=dict(
                   facecolor='red',
                   headwidth=4,
                   width=2,
                   headlength=8,
               ))
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

rect=plt.Rectangle((0.2,0.75),0.4,0.15,color='r',alpha=0.3)#添加常见图形(矩形)
circle=plt.Circle((0.7,0.2),0.15,color='b',alpha=0.5)#添加圆形
ax.add_patch(circle)
ax.add_patch(rect)
```

## 9.2使用pandas和seaborn绘图
### 9.2.1折线图
Series和DataFrame都有一个plot属性,用于绘制基本的图形。默认情况下，plot绘制的是折线图。

```python
import pandas as pd
fig=plt.figure(figsize=(6,9))
ax=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)
series=pd.Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
dataframe=pd.DataFrame(np.random.randn(10,4).cumsum(0),columns=['A','B','C','D'])
series.plot(ax=ax,title='series')
dataframe.plot(ax=ax1)
'''
Series.plot()参数设置:
label     图例标签
ax        绘图所用子图对象
xticks    用于x轴刻度的值
yticks    用于y轴刻度的值
xlim      x轴范围
ylim      y轴范围

DataFrame.plot()参数设置：
subplots     将DataFrame的每一列绘制在独立的列表中
figsize     用于生成图片尺寸的元组
title       标题字符串
'''
```

```python
dataframe=pd.DataFrame(np.random.randn(10,4).cumsum(0),columns=['A','B','C','D'])
dataframe.plot(subplots=True,figsize=(5,20))
```

### 9.2.2柱状图
plot.bar()和plot.barh()可以分别绘制垂直和水平的柱状图。在绘制柱状图的过程中，Series和DataFrame的索引将会被用作x轴刻度(bar)或y轴刻度（barh）。

```python
fig,ax=plt.subplots(1,2)
series=pd.Series(np.random.rand(16),index=list('asdfghjklqwertyu'))
series.plot.bar(ax=ax[0])
df=pd.DataFrame(np.random.rand(6,2),columns=['a','b'])
df.plot.bar(ax=ax[1])
```
